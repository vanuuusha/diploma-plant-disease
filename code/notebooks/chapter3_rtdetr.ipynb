````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display

ROOT = Path('/home/vanusha/diplom/diploma-plant-disease')
TASK_DIR = ROOT / 'code/results/task_08'
VARIANTS = ['baseline', 'aug_geom', 'aug_oversample', 'aug_diffusion']
LOG_FILE = Path('/tmp/train_rtdetr.log')
````

## Живой хвост лога

````python
if LOG_FILE.exists():
    print('\n'.join(LOG_FILE.read_text(errors='ignore').splitlines()[-40:]))
else:
    print('не запущено')
````

## Метрики

````python
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
for v in VARIANTS:
    p = TASK_DIR / f'rtdetr_{v}' / 'metrics.csv'
    if not p.exists():
        continue
    df = pd.read_csv(p)
    axs[0].plot(df['epoch'], df['val_mAP50'], label=v)
    axs[1].plot(df['epoch'], df['val_mAP5095'], label=v)
for ax, t in zip(axs, ['val mAP@50', 'val mAP@50-95']):
    ax.set_xlabel('epoch'); ax.set_ylabel(t); ax.set_title(t); ax.grid(True, alpha=0.3); ax.legend()
plt.tight_layout(); plt.show()
````

## Сводная таблица

````python
p = TASK_DIR / 'summary.csv'
if p.exists():
    display(pd.read_csv(p))
else:
    print('ещё нет')
````

````python
for v in VARIANTS:
    lc = TASK_DIR / f'rtdetr_{v}' / 'learning_curves.png'
    cm = TASK_DIR / f'rtdetr_{v}' / 'confusion_matrix.png'
    print(f'=== {v} ===')
    if lc.exists(): display(Image(str(lc)))
    if cm.exists(): display(Image(str(cm)))
````
